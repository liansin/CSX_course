# HW

觀察到11月份量最大，將其分離出來

```python
a = df[(df['new_DateId'] <= '2017-11-30') & (df['new_DateId'] >= '2017-11-01')]
a
```

觀察是否跟折扣有關

```python
aa = pd.DataFrame(a.groupby(by='new_DateId')['Quantity','EcouponDiscount'].sum())
aa
```

依照銷售數量排列

```python
aa.sort_values(by='Quantity', ascending=False)
```

可以發現11/11光棍節折扣最高，刺激買氣
